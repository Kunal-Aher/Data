```python
from pyspark.sql import SparkSession

spark=SparkSession.builder.appName('dataframe').getOrCreate()
spark
```

```python
#pyspark Group by and Aggregate Function
# create SparkSession
# read dataframe in Spark

from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('Agg').getOrCreate()
spark
```

```python
df=spark.read.option("header","true").csv("/content/sample_data/california_housing_train.csv",inferSchema=True)
df.show()
```

```python
#groupby
gp=df.groupBy('population').sum()
gp.show()
gp1=df.groupBy('population').max().show()
gp1=df.groupBy('population').mean().show()
gp1=df.groupBy('population').count().show()
```
